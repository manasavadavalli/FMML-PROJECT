### Gradient Descent

````python
# Example function and its derivative
def f(x):
    return x**2 - 4*x + 4  # A simple quadratic function

def df(x):
    return 2*x - 4  # Derivative of f(x)

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x

    # Calculate the gradient (derivative) at the current point
    gradient = df(new_x)

    # Update new_x by taking a step in the direction of the negative gradient
    new_x = previous_x - step_multiplier * gradient

    # Calculate the size of the step taken
    step_size = abs(new_x - previous_x)

    # Append the new x value and its corresponding gradient to the lists
    x_list.append(new_x)
    slope_list.append(df(new_x))

    # If the step size is smaller than the precision, break the loop
    if step_size < precision:
        print('Loop ran this many times:', n)
        break

# Output the results after convergence
print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python
## TASK-7 : Plot reduction in cost for each iteration
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def g(x):
    return x**2 - 3*x + 2  # Example function (cost function)

def dg(x):
    return 2*x - 3  # Derivative of the function g(x)

# Gradient Descent Function (tracking cost at each iteration)
def gradient_descent_with_cost(initial_x, gamma, precision=0.00001, max_iter=500):
    new_x = initial_x
    previous_x = 0
    x_list = [new_x]
    cost_list = [g(new_x)]  # Track cost function value at each iteration

    for n in range(max_iter):
        previous_x = new_x
        gradient = dg(new_x)  # Calculate the gradient at current point
        new_x = previous_x - gamma * gradient  # Update new_x using gradient descent

        # Track the cost at the current x value
        current_cost = g(new_x)
        cost_list.append(current_cost)

        step_size = abs(new_x - previous_x)  # Check how much the value changed

        if step_size < precision:  # If the change is small, break the loop
            print(f'Gradient Descent converged in {n} iterations with gamma={gamma}')
            break

    return x_list, cost_list, new_x

# Run Gradient Descent with different learning rates
initial_x = 3  # Starting point for gradient descent
learning_rate = 0.001  # Mid-range learning rate

# Perform gradient descent
x_path, cost_path, final_x = gradient_descent_with_cost(initial_x, learning_rate)

# Step 2: Plot the reduction in cost for each iteration
plt.figure(figsize=(10, 6))

plt.plot(cost_path, label=f'Learning rate: {learning_rate}, Final x: {final_x:.4f}', color='b')
plt.title('Cost Function Reduction During Gradient Descent')
plt.xlabel('Iteration')
plt.ylabel('Cost (g(x))')
plt.legend()
plt.grid(True)
plt.show()
````

````python
import numpy as np
from sympy import symbols, diff

# Define symbols
a, b = symbols('a b')

# Define the cost function f(a, b)
f = a**2 + b**2  # Example function; replace with your actual cost function

# Gradient descent parameters
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])  # Initial guess

for n in range(max_iter):
    # Compute gradients
    gradient_x = diff(f, a).evalf(subs={a: params[0], b: params[1]})
    gradient_y = diff(f, b).evalf(subs={a: params[0], b: params[1]})
    gradients = np.array([gradient_x, gradient_y], dtype=float)

    # Update parameters
    params = params - multiplier * gradients

# Results
print('Values in gradient array:', gradients)
print('Minimum occurs at x value of:', params[0])
print('Minimum occurs at y value of:', params[1])
print('The cost is:', f.evalf(subs={a: params[0], b: params[1]}))
````

````python
from sympy import symbols, diff, lambdify

# Define the variables
x, y = symbols('x y')

# Define the function f(x, y)
f = x**2 + y**2  # Example: Replace with your function

# Compute the partial derivatives symbolically
partial_x = diff(f, x)  # ∂f/∂x
partial_y = diff(f, y)  # ∂f/∂y

# Convert the symbolic derivatives to Python functions
def fpx(x_val, y_val):
    return lambdify((x, y), partial_x)(x_val, y_val)  # Evaluates ∂f/∂x at (x_val, y_val)

def fpy(x_val, y_val):
    return lambdify((x, y), partial_y)(x_val, y_val)  # Evaluates ∂f/∂y at (x_val, y_val)

# Example usage
x_val, y_val = 1.8, 1.0
print("∂f/∂x at (x, y) = (1.8, 1.0):", fpx(x_val, y_val))
print("∂f/∂y at (x, y) = (1.8, 1.0):", fpy(x_val, y_val))
````

````python
import numpy as np
from sympy import symbols, diff, lambdify

# Define symbols
x, y = symbols('x y')

# Define the cost function
f = x**2 + y**2  # Example function; replace this with your actual cost function

# Compute partial derivatives
partial_x = diff(f, x)
partial_y = diff(f, y)

# Convert symbolic expressions to Python functions
f_lambdified = lambdify((x, y), f)  # Converts f(x, y) to a callable Python function
fpx = lambdify((x, y), partial_x)   # Converts ∂f/∂x to a callable Python function
fpy = lambdify((x, y), partial_y)   # Converts ∂f/∂y to a callable Python function

# Gradient descent parameters
multiplier = 0.1
max_iter = 200
params = np.array([1.8, 1.0])  # Initial guess
values_array = params.reshape(1, 2)  # Track parameter values

print("Initial values_array shape:", values_array.shape)

# Gradient descent loop
for n in range(max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients
    values_array = np.concatenate((values_array, params.reshape(1, 2)), axis=0)

# Results
print('Values in gradient array:', gradients)
print('Minimum occurs at x value of:', params[0])
print('Minimum occurs at y value of:', params[1])
print('The cost is:', f_lambdified(params[0], params[1]))
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
````
